```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import requests
from bs4 import BeautifulSoup
import os
import pandas as pd
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException,ElementNotInteractableException,InvalidArgumentException,InvalidCookieDomainException
```

```python
driver = webdriver.Chrome(executable_path="D:\\chromedriver_win32\\chromedriver.exe")
```

```python
url="https://www.naukri.com/"
```

```python
driver.get(url)
```

```python
recruiters=driver.find_element_by_xpath("//a[@title='Search Recruiters']")
driver.get(recruiters.get_attribute('href'))
```

```python
search=driver.find_element_by_xpath("//input[@class='sugInp']")
search.send_keys("Data Science")
```

```python
button=driver.find_element_by_xpath("//button[@id='qsbFormBtn']")
button.click()
```

```python
name=[]
try:
    names=driver.find_elements_by_xpath("//span[@class='fl ellipsis']")
    for i in names:
        name.append(i.text)
except NoSuchElementException:#handling no such element exception
    name.append('No details available')
print(len(name))
```

```python
name
```

```python
designation=[]
try:
    designation_list=driver.find_elements_by_xpath("//span[@class='ellipsis clr']")
    for i in designation_list:
        designation.append(i.text)
except NoSuchElementException:#handling no such element exception
    designation.append('No details available')
print(len(designation))
```

```python
designation
```

```python
company=[]
try:
    company_list=driver.find_elements_by_xpath("//p[@class='highlightable']/a[2]")
    for i in company_list:
        company.append(i.text)
except NoSuchElementException:#handling no such element exception
    company.append('No details available')
print(len(company))
```

```python
company
```

```python
skills=[]
try:
    skills_list=driver.find_elements_by_xpath("//div[@class='hireSec highlightable']")
    for i in skills_list:
        skills.append(i.text)
except NoSuchElementException:#handling no such element exception
    skills.append('No details available')
print(len(skills))
```

```python
skills
```

```python
location=[]
try:
    location_list=driver.find_elements_by_xpath("//small[@class='ellipsis']")
    for i in location_list:
        location.append(i.text)
except NoSuchElementException:#handling no such element exception
    location.append('No details available')

location.insert(29, '-')
location.insert(48, '-')


print(len(location))
```

```python
location
```

```python
Recruters=pd.DataFrame({})
Recruters['Name']=name
Recruters['Designation']=designation
Recruters['Company']=company
Recruters['Skills']=skills
Recruters['Location']=location
```

```python
Recruters
```

```python
Recruters.to_csv("Recruters.csv")
```
